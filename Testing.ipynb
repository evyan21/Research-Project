```
from fastai.vision import *
from PIL import Image, ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
np.random.seed(42)

import numpy as np
```

```
path = '/content/drive/My Drive/Colab Notebooks/wikiart/'
styleorcontent = 'style reference/'
final_path = path+styleorcontent
file_name_list = ["Abstract_Expressionism", "Art_Nouveau_Modern", "Baroque", "Color_Field_Painting", "Cubism", "Early_Renaissance", "Expressionism", "Fauvism", "High_Renaissance", "Impressionism", "Mannerism_Late_Renaissance", "Minimalism", "Naive_Art_Primitivism", "Northern_Renaissance", "Pop_Art", "Post_Impressionism", "Realism", "Rococo", "Romanticism", "Ukiyo_e"]
```

```
# verify test set is 3800
len(ImageList.from_folder(final_path))
```

```
matrix = np.zeros((20, 20))
```

```
#alternative method??



for style in range(len(file_name_list)):
  os.chdir(final_path)
  for content in range(len(file_name_list)):
    #maybe change
    os.chdir(final_path+file_name_list[style])
    if not (style == content):
      # be sure change style and content places
      new_file = file_name_list[style]+"|"+file_name_list[content]
      # change here
      test = ImageList.from_folder(final_path + file_name_list[style] + "/" +new_file)
      final = load_learner(path+'test/models/', test = test)

      #print(len(test))

      # predictions
      preds, y = final.get_preds(ds_type=DatasetType.Test)
      y = torch.argmax(preds, dim=1)

      array_of_ten = y.data.numpy()
      #print(array_of_ten)

      # be sure
      number_correct = (array_of_ten == content).sum()
      # print(number_correct)

      # keep same
      matrix[style, content]=number_correct
    # diagonal spacing
    else:
      matrix[style, content]=0
    #go up one folder

    os.chdir(final_path)
  #reset to home dir

  print(np.array2string(np.divide(matrix[style].T, 10), separator=', '))
  os.chdir(final_path) 
```
