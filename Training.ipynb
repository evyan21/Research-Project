```
from fastai.vision import *
```

```
from PIL import Image, ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```
# initialize data bunch for the classifier
np.random.seed(42)
data = ImageDataBunch.from_folder("/content/drive/My Drive/Colab Notebooks/wikiart/test", train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```
# verify everything is correct
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

```
# create pretrained resnet34 classifer
learn = cnn_learner(data, models.resnet34, metrics=error_rate, pretrained=True)
```

```
learn.load('new_stage-1')
```

```
interp = ClassificationInterpretation.from_learner(learn)
```

```
# con
interp.plot_confusion_matrix(title='Confusion matrix', figsize = (10, 10))
```
